```python
import os
import glob
import pandas as pd
os.chdir("citi_summer_data_2017")
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
all_filenames
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
```

```python
combined_csv.to_csv( "combined_city_2017.csv", index=False, encoding='utf-8-sig')
```

```python
combined_csv.head()
```

```python
df_2017 = combined_csv.drop(columns=['stoptime', 'start station id', 'end station id', 'end station name', 'end station latitude', 'end station longitude', 'bikeid'])
```
