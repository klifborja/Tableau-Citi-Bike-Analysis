```python
import os
import glob
import pandas as pd
os.chdir("../citi_summer_data_2019")
```

```python
extension3 = 'csv'
all_filenames3 = [i for i in glob.glob('*.{}'.format(extension3))]
```

```python
all_filenames3
```

```python
combined_csv3 = pd.concat([pd.read_csv(f) for f in all_filenames3 ])
```

```python
combined_csv3.to_csv( "combined_city_2019.csv", index=False, encoding='utf-8-sig')
```

```python
combined_csv3.head()
```

```python
df_2019 = combined_csv3.drop(columns=['stoptime', 'end station name', 'end station latitude', 'end station longitude'])
```
