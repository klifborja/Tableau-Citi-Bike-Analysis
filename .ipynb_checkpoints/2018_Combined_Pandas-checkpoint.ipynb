```python
import os
import glob
import pandas as pd
os.chdir("citi_summer_data_2018")
```

```python
extension2 = 'csv'
all_filenames2 = [i for i in glob.glob('*.{}'.format(extension2))]
```

```python
all_filenames2
```

```python
combined_csv2 = pd.concat([pd.read_csv(f) for f in all_filenames2 ])

combined_csv2.to_csv( "combined_city_2018.csv", index=False, encoding='utf-8-sig')
```

```python
combined_csv2.head()
```

```python
df_2018 = combined_csv2.drop(columns=['stoptime', 'start station id', 'end station id', 'end station name', 'end station latitude', 'end station longitude', 'bikeid'])
```
